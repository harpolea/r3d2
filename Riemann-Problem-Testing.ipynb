# Test bench problems

See the [Living Review of Mart and Mller](http://computastrophys.livingreviews.org/Articles/lrca-2015-3/articlese6.html#x9-690006.3) for this particular set of problems.

```python
eos = eos_defns.eos_gamma_law(5.0/3.0)
test_1_w_left = SR1d.State(10.0, 0.0, 0.0, 2.0, eos, label="L")
test_1_w_right = SR1d.State(1.0, 0.0, 0.0, 1.5e-6, eos, label="R")
test_1_rp = SR1d.RP(test_1_w_left, test_1_w_right)
test_2_w_left = SR1d.State(1.0, 0.0, 0.0, 1500.0, eos, label="L")
test_2_w_right = SR1d.State(1.0, 0.0, 0.0, 1.5e-2, eos, label="R")
test_2_rp = SR1d.RP(test_2_w_left, test_2_w_right)
test_3_w_left = SR1d.State(1.0, 0.0, 0.0, 1500, eos, label="L")
test_3_w_right = SR1d.State(1.0, 0.0, 0.99, 0.015, eos, label="R")
test_3_rp = SR1d.RP(test_3_w_left, test_3_w_right)
test_4_w_left = SR1d.State(1.0, 0.0, 0.9, 1500, eos, label="L")
test_4_w_right = SR1d.State(1.0, 0.0, 0.9, 0.015, eos, label="R")
test_4_rp = SR1d.RP(test_4_w_left, test_4_w_right);
```

# More general problems

The equation of state used doesn't have to be the same on both sides of the interface. This example corresponds to figure 3 in [Millmore & Hawke](http://dx.doi.org/10.1088/0264-9381/27/1/015007), where two different gamma-law equations of state are used.

```python
eos1 = eos_defns.eos_gamma_law(1.4)
eos2 = eos_defns.eos_gamma_law(1.67)
test_multi_eos_w_left = SR1d.State(10.2384, 0.9411, 0.0, 50.0/0.4/10.23841, eos1, label="L")
test_multi_eos_w_right = SR1d.State(0.1379, 0.0, 0.0, 1.0/0.1379/0.67, eos2, label="R")
test_multi_eos_rp = SR1d.RP(test_multi_eos_w_left, test_multi_eos_w_right);
```

```python
display(test_multi_eos_rp)
display_svg(test_multi_eos_rp)
```
